```python
# we don't have any viability data for this one
viab1 = pd.DataFrame({'cell_line':[], 'conc_nm':[], 'inhibitor':[], 'cell_viab':[]})
viab1.head()
```

# live cell reporter imaging 

```python
clover = pd.read_csv('../data/HEK8/clover_all_cell.csv')
print('shape', clover.shape)
print('unique tracks', clover.track_index.unique().shape)
clover.head()
```

```python
mscarlet = pd.read_csv('../data/HEK8/mscarlet_all_cell.csv')
print('shape', mscarlet.shape)
print('unique tracks', mscarlet.track_index.unique().shape)
mscarlet.head()
```

```python
# grab time 
time = clover.columns[1:-4]
print(time)

# create new time designators + selectors for later 
clover_selector = [f'cdk2_time(h)={x}' for x in time]
mscarlet_selector = [f'mdmx_time(h)={x}' for x in time]

# update time cols 
clover = clover.rename({x:f'cdk2_time(h)={x}' for x in time}, axis=1)
mscarlet = mscarlet.rename({x:f'mdmx_time(h)={x}' for x in time}, axis=1)
clover = clover.rename({'median':'cdk2_median', 'site':'cdk2_site'}, axis=1)
mscarlet = mscarlet.rename({'median':'mdmx_median', 'site':'mdmx_site'}, axis=1)
```

```python
# merge clover and mscarlet
data = clover.merge(mscarlet, on=['track_index', 'cell__treatment'])

print('shape', data.shape)
print('unique tracks', data.track_index.unique().shape)
```

## create `conc_nm` and `inhibtor` columns

```python
# eda 
data.groupby('cell__treatment')['track_index'].count()
```

[3:49 PM] Samuel Tsang  
    5um trametinib, 1um bmn673, 1um etoposide, 5um trametinib + 1um bmn673

```python
treatment_map = {'bmn673':(1e3,'bmn673'),
 'bmn673+trametinib':((1e3,5e3),'bmn673,trametinib'), 
 'etoposide':(1e3,'etoposide'),
 'trametinib':(5e3,'trametinib'), 
 'untreated':(0,'untreated')}
```

```python
data = data.assign(conc_nm=[treatment_map[x.split('__')[-1]][0] for x in data.track_index],
                  inhibitor=[treatment_map[x.split('__')[-1]][1] for x in data.track_index],
                  cell_line='heya8')
```

## merge viability data 

```python
data = data.merge(viab1, on=['cell_line', 'inhibitor', 'conc_nm'], how='left', validate='m:1')

print('shape', data.shape)
print('unique tracks', data.track_index.unique().shape)
```

```python
# set cell viab nan's to 1 - only nans should be no treatment
#data.cell_viab = data.cell_viab.fillna(1.)
data.head()
```

# Add metadata to `dataframe` and save as `.pkl` 

```python
data.attrs['clover_selector'] = clover_selector 
data.attrs['mscarlet_selector'] = mscarlet_selector
data.attrs['treatment_map'] = treatment_map
data.attrs['meta'] = {'pathways':('cdk2','mdmx'), 'cell_line':'heya8', 'treatments':('trametinib','bmn673','etoposide')}
```

```python
with open('../data/HEK8.pkl', 'wb') as f: 
    pkl.dump(data, f)
```
