```python
import numpy as np 
import pandas as pd
from hmmlearn import hmm

from matplotlib import pyplot as plt
import seaborn as sbn
```

# Overview 


HMM using [this](https://hmmlearn.readthedocs.io/en/latest/) package. 

```python
dat = pd.read_csv('../data/clover_all_cell.csv')
dat.head()
```

```python
# https://hmmlearn.readthedocs.io/en/latest/tutorial.html#multiple-sequences

X = dat[dat.columns[1:-3]].values.reshape(-1,1)
_len = np.array([len(dat.columns[1:-3])]*dat.shape[1])
```

```python
X.shape
```

```python
_len
```

```python
?hmm.GaussianHMM
```

```python
remodel = hmm.GaussianHMM(n_components=3, covariance_type="full", n_iter=100, init_params='stc',)
remodel.means_ = np.array([[0.0], [0.5], [1]]) # ensure that our states are ordered low to high (e.g., 0=low, 1=mid...)
remodel.fit(X)
Z2 = remodel.predict(X)
```

```python
remodel.monitor_.converged
```

```python
z2 = Z2.reshape((dat.shape[0], len(dat.columns[1:-3])))
```

```python
remodel.means_
```

```python
plt.figure(figsize=(20,10))
_n = 500
plt.plot(X[:_n], 'r', label='raw')
plt.plot(Z2[:_n]/2, 'b', label='predicted state')
plt.legend()

plt.show()
```

```python
res = pd.DataFrame(z2, columns=dat.columns[1:-3])
res = res.assign(label=dat.track_index)
res = res.set_index('label').unstack().reset_index().rename({'level_0':'time', 0:'state'}, axis=1)
res.head()
```

```python
out = res.groupby(['label','state']).count().rename({'time':'count'},axis=1).reset_index().set_index('label').pivot(columns='state', values='count').reset_index()
out = out.fillna(0)
out = out.merge(dat[['track_index', 'cell__treatment']], left_on='label', right_on='track_index')
out = out.rename({0:'c0', 1:'c1', 2:'c2'}, axis=1)
out.head()
```

```python
plt.figure(figsize=(10,10))
sbn.scatterplot(x='c0', y='c1', hue='cell__treatment', data=out)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sbn.scatterplot(x='c0', y='c2', hue='cell__treatment', data=out)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sbn.scatterplot(x='c1', y='c2', hue='cell__treatment', data=out)
plt.show()
```
